```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
(xtrain,ytrain),(xtest,ytest)=keras.datasets.mnist.load_data()
```

```python
xtrain[0].shape
```

```python
print("the xtraining test shape is ",xtrain.shape)
print("the xtesting shape is",xtest.shape)
print("the ytraining shape is",ytrain.shape)
print("the ytest shape is ",ytest.shape)
```

```python
# scaling technique
xtrain=xtrain.reshape(len(xtrain),28*28)
print(xtrain.shape)
xtrainscaled=xtrain/255
```

```python
def image(index):
 plt.matshow(xtrain[index])
 return index
```

```python
image(2)
```

```python
ytrain[2]
```

```python
# implement neural network
```

```python
model=keras.Sequential([
    keras.layers.Dense(100,input_shape=(784,),activation="relu"),
    keras.layers.Dense(10,activation="sigmoid")
])
model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"]
             )
```

```python
model.fit(xtrainscaled,ytrain,epochs=10)
```

```python
model.evaluate(xtrainscaled,ytrain)
```

```python
pred=model.predict(xtrainscaled)
pred=[np.argmax(i) for i in pred]
pred
```

```python
np.argmax(pred[2])
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
confusion_matrix=tf.math.confusion_matrix(labels=ytrain,predictions=pred)
```
