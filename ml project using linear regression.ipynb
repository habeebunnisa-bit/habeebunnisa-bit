```python
df=pd.read_csv(r'C:\Users\RINNY\Downloads\archive (4)\heart.csv')
df
```

```python
from sklearn.model_selection import  train_test_split
x=df.drop(columns='target')
y=df['target']
```

```python
xtrain,ytrain,xtest,ytest=train_test_split(x,y,test_size=0.20,random_state=45)
print("the xtrain shape is",xtrain.shape)
print("the xtest shape is",xtest.shape)
print("the ytrain shape is",ytrain.shape)
print("the ytest shape is",ytest.shape)
```

```python
#linear regression model 
linear_model=LinearRegression()
linear_model.fit(x,y)
```

```python
linear_model.coef_
```

```python
linear_model.intercept_
```

```python
pred=linear_model.predict(xtrain)
prediction=[1 if i<1.0 else 0 for i in pred]
prediction[5]
```

```python
import tensorflow as tf
import seaborn as sns
from sklearn.metrics import confusion_matrix,classification_report
confusion_matrix=tf.math.confusion_matrix(labels=xtest,predictions=prediction)
sns.heatmap(confusion_matrix,annot=True,fmt='d')
```

```python
linear_model.score(xtrain,xtest)
```

```python
linear_model.score(ytrain,ytest)
```
