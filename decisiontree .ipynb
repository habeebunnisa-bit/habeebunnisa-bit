```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
data=pd.read_csv(r'C:\Users\RINNY\Downloads\winequality-red.csv')
data
```

```python
x=data.drop(columns='quality')
x
y=data['quality']
y
```

```python
#scaling data 
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
x_scalar=scalar.fit_transform(x)
x_scalar
```

```python
#training  and testing data
xtrain,ytrain,xtest,ytest=train_test_split(x_scalar,y,test_size=0.25,random_state=45)
print("the xtrain data is",xtrain)
print("the ytrain data is ",ytrain)
print("the xtest data is ",xtest)
print("the ytest data is ",ytest)
```

```python
decisiontree_model=DecisionTreeClassifier(criterion='entropy',splitter='best',random_state=45,ccp_alpha=0.9)
decisiontree_model
```

```python
model=decisiontree_model.fit(xtrain,xtest)
model
```

```python
from sklearn import tree
plt.figure(figsize=(5,5))
tree.plot_tree(model,filled=True,feature_names=x.columns)
```

```python
pred=model.predict(xtrain)
pred
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import tensorflow as tf
confusion_matrix=tf.math.confusion_matrix(labels=xtest, predictions=pred)
sns.heatmap(confusion_matrix)
```
