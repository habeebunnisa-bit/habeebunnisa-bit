```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv(r'C:\Users\RINNY\Downloads\homeprices.csv')
print(data)
```

```python
x = data[['area', 'bedrooms', 'age']]
y = data['price']
print(x)
print(y)
```

```python
scalar=StandardScaler()
scalar
```

```python
x_scalar=scalar.fit_transform(x)
print(x_scalar)
```

```python
from sklearn.model_selection import train_test_split
xtrain,ytrain,xtest,ytest=train_test_split(x_scalar,y,random_state=45)
print("the xtrain is",xtrain)
print("the ytrain is ",ytrain)
print("the xtest is ",xtest)
print("the ytest is ",ytest)
```

```python
linear_regression_model=LinearRegression()
linear_regression_model
```

```python
linear_regression_model.fit(xtrain,xtest)
```

```python
linear_regression_model.coef_
```

```python
linear_regression_model.intercept_
```

```python
pred=linear_regression_model.predict(xtrain)
prediction=np.abs(pred)
prediction
```
